```python
cols = ['1990 [YR1990]', '2000 [YR2000]', '2014 [YR2014]',
       '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]',
       '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]', '2022 [YR2022]',
       '2023 [YR2023]']
cols
```

```python
for i in cols:
    fig = plt.figure(figsize=(6,6))
    plt.hist(df_cleaned[i],bins=10, color='blue')
    plt.xlabel(i)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {i}')
    plt.show()
```

```python
for n in cols:
    df_cleaned[n] = pd.to_numeric(df_cleaned[n], errors='coerce')

column_to_visualize = '2015 [YR2015]'
df_cleaned = df_cleaned.dropna(subset=[column_to_visualize])
plt.figure(figsize=(20,16))
plt.bar(df_cleaned['Country Name'], df_cleaned[column_to_visualize], color='blue')
plt.xlabel('Country Name')
plt.ylabel(column_to_visualize)
plt.title(f'Distribution of population in {column_to_visualize}')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
