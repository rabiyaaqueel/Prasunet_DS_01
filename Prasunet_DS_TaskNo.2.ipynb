```python
# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace = True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace = True)

# Drop the 'Cabin' column
df.drop(columns=['Cabin'], inplace=True)
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
#Univariate Analysis
# Age distribution
plt.figure(figsize =(20,10))
sns.histplot(df['Age'], kde=True, color='Red')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Univariate Analysis
#Fare Diatribution
plt.figure(figsize=(20,10))
sns.histplot(df['Fare'], kde=True, color='Red')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
#Univariate Analysis
#Survival Rate
plt.figure(figsize=(20,10))
sns.countplot(x='Survived', data=df, palette='dark')
plt.title('Survival count')
plt.xlabel('Survival')
plt.ylabel('Count')
plt.show()
```

```python
#Bivariate Analysis

#Survival by sex
plt.figure(figsize=(20,10))
sns.countplot(x='Survived', hue='Sex', data=df, palette='dark')
plt.title('Survival by sex')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='Sex', loc='upper right', labels=['Male', 'Female'])
plt.show()
```

```python
#Bivariate Analysis

#Survival by passenger class
plt.figure(figsize=(20,10))
sns.countplot(x='Survived', hue='Pclass', data=df, palette='dark')
plt.title('Survival by passenger class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='Pclass', loc='upper right', labels=['1st', '2nd', '3rd'])
plt.show()
```

```python
#Bivariate Analysis

#Age distribution by survival
plt.figure(figsize=(20,10))
sns.countplot(data=df, x='Age', hue='Survived', fill=True, palette='dark')
plt.title('Age distribution by survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
#Multivariate Analysis

#pairplot
plt.figure(figsize=(20,10))
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Sex', 'Pclass']], hue='Survived', palette='dark')
plt.show()
```

```python
#Correlation Matrix
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(), annot=True, cmap='Blues', linewidths=0.12)
plt.title("Correlation Matrix")
plt.show()
```

```python
#Correlation Matrix

#Define Age Groups
bins=[0, 12, 18, 30, 60, 100]
labels=['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior']
df['Age group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Survival by Age group
plt.figure(figsize=(10, 6))
sns.countplot(x='Age group', hue='Survived', data=df, palette='Blues')
plt.title('Survival by Age group')
plt.xlabel('Age group')
plt.ylabel('Count')
```
